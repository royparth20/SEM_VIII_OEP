```python
import copy
class TicTacToe:
    def __init__(self):
        self.player = None
        self.board = None
        self.opponent = None
        self.bestMove = None

    def set_board(self, board):
        self.board = board

    def set_player(self, player):
        self.player = player

    def set_opponent(self, opponent):
        self.opponent = opponent

    def evaluate(self, b):
        for row in range(0, 3):
            if b[row][0] == b[row][1] and b[row][1] == b[row][2]:

                if b[row][0] == self.player:
                    return 10
                elif b[row][0] == self.opponent:
                    return -10

        # Checking for Columns for X or O victory.
        for col in range(0, 3):

            if b[0][col] == b[1][col] and b[1][col] == b[2][col]:

                if b[0][col] == self.player:
                    return 10
                elif b[0][col] == self.opponent:
                    return -10
        if b[0][0] == b[1][1] and b[1][1] == b[2][2]:

            if b[0][0] == self.player:
                return 10
            elif b[0][0] == self.opponent:
                return -10

        if b[0][2] == b[1][1] and b[1][1] == b[2][0]:

            if b[0][2] == self.player:
                return 10
            elif b[0][2] == self.opponent:
                return -10
        return 0

    def isFullBoard(self, board):
        for x in board:
            for y in x:
                if y == '':
                    return False
        return True

    def createNewNode(self, currentBoard, nChar, oChar):
        newNode = currentBoard[:]
        c = 0
        for x in currentBoard:
            col = 0
            for y in x:
                if y == oChar:
                    newNode[c][col] = nChar
                    return newNode
                col -= -1
            c -= -1
        return newNode

    def minimax(self, board, depth, isMax):
        score = self.evaluate(board)
        self.depth=depth
        print("Score : ", score, "\nDepth : ", depth, "\nMovement : ", isMax)

        if score == 10:
            return score,depth
        if score == -10:
            return score,depth

        if self.isFullBoard(board):
            return 0,depth

        if isMax:
            best = -999
            temp = self.createNewNode(board, self.player, '')
            # print(temp)
            nbest,ndepth=self.minimax(temp, depth+1,
                                          not isMax)
            best = max(best, nbest)
            self.depth=max(self.depth,ndepth)
#             print('Max :', best)
            return best,depth
        
        else:
            best = 999
            temp = self.createNewNode(board, self.opponent, '')
            # print(temp)
            nbest,ndepth=self.minimax(temp, depth+1,
                                          not isMax)
            best = min(best, nbest)
            self.depth=max(self.depth,ndepth)
#             print('Min :',best)
            return best,depth

    def findNumberEmpty(self):
        emptySpace = []
        row = 0
        for x in self.board:
            col = 0
            for y in x:
                if y == '':
                    emptySpace.append([row, col])
                col -= -1
            row -= -1
        return emptySpace

    def findBestMove(self):
        self.bestVal = -999
        eSpaces = self.findNumberEmpty()
        bestDepth=0
        for i in eSpaces:
            nboard = copy.deepcopy(self.board)
            row=i[0]
            col=i[1]
            temp = 0
            nboard[row][col] = self.player
            print("Input :", nboard)
            self.depth=0
            movVal,ndepth = self.minimax(nboard, 0, False)
            print("Self Depth : ",self.depth,ndepth,bestDepth)
            if movVal >= self.bestVal :
                if movVal==self.bestVal:
                    if bestDepth==self.depth:
                        self.bestMove = row, col
                        self.bestVal = movVal
                    elif bestDepth<self.depth:
                        self.bestMove = row, col
                        self.bestVal = movVal
                else:
                    self.bestMove = row, col
                    self.bestVal = movVal
                print(i[0], i[1],nboard)
                print("-"*10, "BestVal :", self.bestVal, "-"*10)
        return self.bestMove
    def minimaxtemp(self,board,depth,isMax):
        score = self.evaluate(board)
        self.depth=depth
        print("Score : ", score, "\nDepth : ", depth, "\nMovement : ", isMax)
        if score == 10:
            return score
        if score == -10:
            return score
        if self.isFullBoard(board):
            return 0
        if isMax:
            best = -999
            eSpaces=self.findNumberEmpty()
            for i in eSpaces:
                print(i)
            temp = self.createNewNode(board, self.player, '')
            # print(temp)
            nbest,ndepth=self.minimax(temp, depth+1,
                                          not isMax)
            best = max(best, nbest)
            self.depth=max(self.depth,ndepth)
#             print('Max :', best)
            return best
        
        else:
            best = 999
            temp = self.createNewNode(board, self.opponent, '')
            # print(temp)
            nbest,ndepth=self.minimax(temp, depth+1,
                                          not isMax)
            best = min(best, nbest)
            self.depth=max(self.depth,ndepth)
#             print('Min :',best)
            return best

    def findBestMoveTemp(self):
        self.bestval=-999
        eSpaces = self.findNumberEmpty()
        bestDepth=0
        self.keys=[]
        self.values=[]
        for i in eSpaces:
            nboard = copy.deepcopy(self.board)
            row=i[0]
            col=i[1]
            nboard[row][col] = self.player
#             print(row,col,nboard)
            self.values.append(self.board)
            self.keys.append(nboard)
            
        for x in range(len(self.keys)):
            movVal=self.minimaxtemp(self.keys[x],0,False)
            print(movVal)
            print(self.keys[x],"--"*10,">>>>",self.values[x])
        
        pass
        
```
